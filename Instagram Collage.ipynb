```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
Answer = np.zeros((430,430,3),dtype='int')
top_left= cv2.imread('top_left.jpg')
top_left = cv2.resize(top_left,(200,200))
top_right= cv2.imread('top_right.jpg')
top_right = cv2.resize(top_right,(200,200)) 
bottom_left= cv2.imread('bottom_left.jpg')
bottom_left = cv2.resize(bottom_left,(200,200))
bottom_right= cv2.imread('bottom_right.jpg')
bottom_right = cv2.resize(bottom_right,(200,200)) 
Answer[10:210,10:210] = top_left[:,:]
Answer[220:420,220:420] = bottom_right[:,:]
Answer[10:210,220:420] = top_right[:,:]
Answer[220:420,10:210] = bottom_left[:,:]

center = cv2.imread("center.jpeg")
center = cv2.resize(center,(100,100))
Answer[155:275,155:275] = 0
Answer[165:265,165:265] = center[:,:]
Answer = Answer[:,:,::-1]

cv2.imwrite("Final.jpg",Answer[:,:,::-1])
plt.imshow(Answer)

final = np.reshape(Answer,(430*430,3))
df = pd.DataFrame(final)
df.to_csv("Answer.csv",index=False,header=("r","g","b"))




```
